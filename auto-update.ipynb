```python
#將民國轉西元前，處理/
def transfer(report):
    report = report.split("/") #用/分割
    num = ""
    for i in range(0, len(report)):
        num =  num + report[i]
    return num

for i in range(0, len(report)):
    for j in range(0, 9):
        if( "/" in report[i][j]):
            report[i][j] = transfer(report[i][j])
pd.DataFrame(report)
```

```python
#民國轉西元
for j in range(0, len(report)):
    report[j][0] = int(report[j][0])+19110000
    temp = str(report[j][0])
    datetime.datetime.strptime(temp, "%Y%m%d")
pd.DataFrame(report)
```
